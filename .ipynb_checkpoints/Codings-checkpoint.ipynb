#### Data Preprocessing

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('Data analysis.csv', delimiter = ';')
dataset
```

```python
jumlah_data =  len(dataset)
print(f"Jumlah data : {jumlah_data}")
```

```python
import re
import nltk

nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

kalimat = []
for i in range(0, jumlah_data): 
    tulisan = re.sub('[^a-zA-Z]', ' ', dataset['text'][i]) 
    tulisan = tulisan.lower() 
    tulisan = tulisan.split() 
    ps = PorterStemmer()
    kata2 = stopwords.words('english')
    tulisan = [ps.stem(kata) for kata in tulisan if not kata in set(kata2)]
    tulisan = ' '.join(tulisan) 
    kalimat.append(tulisan)

kalimat 
```

```python
kata2
```

#### Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(kalimat).toarray()
y = dataset.iloc[:jumlah_data, 0].values

X
```

#### Data Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

#### Menggunakan model algoritma naive bayes (CONTOH)

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

Confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 =  f1_score(y_test, y_pred)

print(f"\nConfusion Matrix model : Naive bayes") 
print(cm) 
print(f"\nAccuracy : {accuracy:.3f}")
print(f"Precision  : {precision:.3f}")
print(f"Recall     : {recall:.3f}")
print(f"F-Score    : {f1:.3f}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score

# Load dataset
data = pd.read_csv('Data analysis.csv', delimiter = ';')

# Pisahkan fitur dan label
X = data['text']  # Kolom teks
y = data['label']  # Kolom label kategori

# Encoding label (ubah teks ke angka)
y = y.str.lower()  # Normalisasi teks
y = y.map({'normal': 0, 'stress': 1})

# Vektorisasi teks
vectorizer = CountVectorizer(stop_words='english')
X_features = vectorizer.fit_transform(X).toarray()

# Bagi data menjadi train-test
X_train, X_test, y_train, y_test = train_test_split(X_features, y, test_size=0.2, random_state=42)

# Melatih model GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)

# Prediksi
y_pred = model.predict(X_test)

# Evaluasi hasil
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test,y_pred))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score
```

```python
data = pd.read_csv('Data analysis.csv', delimiter = ';')
data
```

```python
X = data['text']  # Kolom teks
y = data['label']  # Kolom label kategori

print("Sample x : ", X.head())
print("Sample y : ", y.head())
```

```python
y = y.str.lower()  # Normalisasi teks
y = y.map({'normal': 0, 'stress': 1})

print("Unique values in y: ", y.unique())
```

```python
vectorizer = CountVectorizer(stop_words='english')
X_features = vectorizer.fit_transform(X).toarray()

print("Shape of X_features: ", X_features.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_features, y, test_size=0.2, random_state=42)
```

```python
model = GaussianNB()
model.fit(X_train, y_train)

# Prediksi
y_pred = model.predict(X_test)

# Evaluasi hasil
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test,y_pred))
```
