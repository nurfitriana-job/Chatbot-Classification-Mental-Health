```python
# --- Import Library Umum ---
import pandas as pd
import numpy as np
import pickle
import nltk
import re

# --- Import Library NLTK ---
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords


# --- Import Library Scikit-learn ---
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# --- Import Library Display ---
from IPython.display import display

nltk.download('stopwords')
```

```python
dataset = pd.read_csv('dataanalysis.csv', delimiter=',')
display(dataset.head())

jumlah_data = len(dataset)
print(f"\nJumlah data: {jumlah_data}")
```

```python
kalimat = []
for i in range(0, jumlah_data):
    tulisan = re.sub('[^a-zA-Z]', ' ', dataset['Tweet'][i])
    tulisan = tulisan.lower()
    tulisan = tulisan.split()
    ps = PorterStemmer()
    kata2 = stopwords.words('indonesian')
    tulisan = [ps.stem(kata) for kata in tulisan if not kata in set(kata2)]
    tulisan = ' '.join(tulisan)
    kalimat.append(tulisan)
```

```python
cv = CountVectorizer(max_features=1500)
X = cv.fit_transform(kalimat).toarray()
y = dataset.iloc[:jumlah_data, 0].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=0
)
print(f"\nData Train: {X_train.shape[0]} samples")
print(f"Data Test: {X_test.shape[0]} samples")
```

```python
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
knn = KNeighborsClassifier()
svm = SVC(probability=True)
lr = LogisticRegression()
nn = MLPClassifier()
nb = GaussianNB()

jenis_model = {
    'Logistic Regression': lr,
    'Neural Network': nn,
    'Decision Tree': dt,
    'Random Forest': rf,
    'Naive Bayes': nb,
    'K-NN': knn,
    'SVM': svm,
}
```

```python
evaluasi = {}
for nama, model in jenis_model.items():
    print(f"\nMelatih model: {nama}")
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)

    evaluasi[nama]={
        'prediksi' : y_pred,
        'nilai_true' : y_test,
        'confusion_matrix' :confusion_matrix (y_test,y_pred),
        'akurasi' : accuracy_score (y_test,y_pred),
        'presisi' : precision_score (y_test,y_pred),
        'recall' : recall_score(y_test,y_pred),
        'f1' : f1_score(y_test,y_pred)
    }

    print(f"Model '{nama}' telah dilatih dan dievaluasi.")
```

```python
df_evaluasi = pd.DataFrame.from_dict(
    evaluasi,
    orient='index'
).sort_values(by='akurasi',
              ascending=False
).reset_index(
).rename(columns={'index': 'Model'})

df_evaluasi
```

```python
for nama, model in jenis_model.items():
    model_data = {
        'vectorizer': cv,
        'model': model
    }
    model_filename = f"{nama.replace(' ', '_').lower()}.pkl"
    with open(model_filename, 'wb') as file:
        pickle.dump(model_data, file)
    print(f"Model '{nama}' dan CountVectorizer berhasil disimpan pada file: '{model_filename}'.")
```

```python
model_filename = 'neural_network.pkl'
with open(model_filename, 'rb') as file:
    model_data = pickle.load(file)
print(f"Model '{model_filename}' berhasil di load.")

cv_loaded = model_data['vectorizer']
model_loaded = model_data['model']

new_tweets = [
    # Sentimen Positif
    "Pelayanan cepat dan ramah membuat pengalaman berbelanja saya sangat menyenangkan.",
    "Produk berkualitas tinggi dengan desain menarik dan pengiriman tepat waktu, sangat memuaskan.",

    # Sentimen Negatif
    "Kualitas produk buruk dan mudah rusak setelah beberapa kali digunakan.",
    "Layanan pelanggan tidak responsif dan proses pengembalian barang sangat rumit.",
    "Pengiriman terlambat tanpa pemberitahuan, layanan pelanggan juga kurang responsif."
]


kalimat_baru = []
for tweet in new_tweets:
    tulisan = re.sub('[^a-zA-Z]', ' ', tweet)
    tulisan = tulisan.lower()
    tulisan = tulisan.split()
    ps = PorterStemmer()
    kata2 = set(stopwords.words('indonesian'))
    tulisan = [ps.stem(kata) for kata in tulisan if kata not in kata2]
    tulisan = ' '.join(tulisan)
    kalimat_baru.append(tulisan)

dot = "\n" + "-" * 55 + "\n"
print(dot)
print("Preprocessing Teks:\n")
for original, processed in zip(new_tweets, kalimat_baru):
    print(f"Before: {original}\nAfter: {processed}\n")

X_new = cv_loaded.transform(kalimat_baru).toarray()
y_new_pred = model_loaded.predict(X_new)

sentiment_mapping = {0: 'negatif',
                     1: 'positif'}

print(dot)
print("Prediksi Sentimen:\n")
for tweet, pred in zip(new_tweets, y_new_pred):
    sentiment = sentiment_mapping[pred]
    print(f"Teks: {tweet}\nPrediksi Sentimen: {sentiment}\n")
```
