### Data Preprocessing

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('Data analysis.csv', delimiter = ';')
dataset
```

```python
jumlah_data =  len(dataset)
print(f"Jumlah data : {jumlah_data}")
```

```python
#memisahkan label dan teks

X = dataset['text']  # Kolom teks
y = dataset['label']  # Kolom label kategori

print("Sample x : ", X.head())
print("Sample y : ", y.head())
```

```python
# Mengubah label yang masih berbentuk teks menjadi angka biner)

y = y.str.lower()  # Normalisasi teks
y = y.map({'normal': 0, 'stress': 1})

print("Values in y: ", y.unique())
```

```python
import re
import nltk

nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

kalimat = []
for i in range(0, jumlah_data): 
    tulisan = re.sub('[^a-zA-Z]', ' ', dataset['text'][i]) 
    tulisan = tulisan.lower() 
    tulisan = tulisan.split() 
    ps = PorterStemmer()
    kata2 = stopwords.words('english')
    tulisan = [ps.stem(kata) for kata in tulisan if not kata in set(kata2)]
    tulisan = ' '.join(tulisan) 
    kalimat.append(tulisan)

kalimat 
```

```python
kata2
```

```python
#Vektorisasi 
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(stop_words='english')
X_features = vectorizer.fit_transform(X).toarray()

print("Shape of X_features: ", X_features.shape)
```

#### Data Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_features, y, test_size=0.2, random_state=42)
```

#### Menggunakan model algoritma decision tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
#Prediksi menggunakan model decision tree
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 =  f1_score(y_test, y_pred)

print(f"\nConfusion Matrix model : Decision Tree") 
print(cm) 
print(f"\nAccuracy : {accuracy:.3f}") 
print(f"Precision  : {precision:.3f}")
print(f"Recall     : {recall:.3f}")
print(f"F-Score    : {f1:.3f}")
```

```python
#visualisasi
import matplotlib.pyplot as plt
labels = ['Accuracy','Precision','Recall','F1-score']
values = [accuracy, precision, recall, f1]
plt.barh(labels, values)
 
for index, value in enumerate(values):
    plt.text(value, index, str("{:.3f}".format(value)))
    
plt.title('Evaluasi Decision Tree Model')
plt.show ()
```
