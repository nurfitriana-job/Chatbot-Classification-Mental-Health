```python
import numpy as np
import pandas as pd

#untuk sentimen analisis
dataset = pd.read_csv('dataanalysis.csv', delimiter=',')
dataset
```

```python
jumlah_data =  len(dataset)
print(f"Jumlah data : {jumlah_data}")
```

### Pembersihan Data

```python
pip install nltk
```

```python
import re
import nltk #dipake untuk bersihin teks

nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

kalimat = []
for i in range(0, jumlah_data): #dibersihin semua data sesuai jumlah data yg udh didapatkan
    tulisan = re.sub('[^a-zA-Z]', ' ', dataset['Tweet'][i]) #diambil cuma huruf aja, simbol ga diambil
    tulisan = tulisan.lower() #ubah huruf kapital jadi huruf kecil
    tulisan = tulisan.split() #membagi kalimat jadi per kata
    ps = PorterStemmer()
    kata2 = stopwords.words('indonesian') #untuk menghapus kata2 yang tidak digunakan atau yg ga penting
    tulisan = [ps.stem(kata) for kata in tulisan if not kata in set(kata2)]
    tulisan = ' '.join(tulisan) #disatuin lagi jadi kalimat kata2 yang udh di filter tadi
    kalimat.append(tulisan)

kalimat #hasil output kalimat disini yang udh dibersihin kata2 ga pentingnya
```

```python
kata2 #ini kata2 yang dihapus oleh stopwords
```

### Vectorizer

```python
pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(kalimat).toarray()
y = dataset.iloc[:jumlah_data, 0].values

X
```

### Data Splitting (pembagian data)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)

#data testing utk ngetest model machine learning setelah diajarkan di training
#data training utk latih model machine learning
```

### Pertemuan 3

```python
#menggunakan model algoritma naive bayes

from sklearn.naive_bayes import GaussianNB
GaussianClassifier = GaussianNB()
GaussianClassifier.fit(X_train, y_train)
```

```python
#Membuat dan menampilkan prediksi model algoritma naive bayes
#hasil prediksinya [1,1] = True positif [0,1] = false positif, [1,0] = true negative, [0,0] = false negative 

y_pred = GaussianClassifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#Membuat dan menampilkan confusion matrix model algoritma naive bayes

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 =  f1_score(y_test, y_pred)

print(f"\nConfusion Matrix model : Naive bayes") #kaitannya ke true false - positive negative
print(cm) #hasil matriks disini beda2 karena sesuai sama spek laptop
print(f"\nAccuracy : {accuracy:.3f}") #3f artinya 3 angka belakang koma
print(f"Precision  : {precision:.3f}")
print(f"Recall     : {recall:.3f}")
print(f"F-Score    : {f1:.3f}")
```

```python
#Membuat prediksi dan confusion matrix banyak model algoritma
#matrix
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

#models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

nb = GaussianNB()
lr = LogisticRegression()
knn = KNeighborsClassifier()
svm = SVC (probability=True)
nn = MLPClassifier()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()

jenis_model = {
    'Naive Bayes' : nb,
    'Logistic Regression' : lr,
    'K-NN' : knn,
    'SVM' : svm,
    'Neural Network' : nn,
    'Decission Tree' : dt,
    'Random Forest' : rf,
}

evaluasi = {}
for nama, model in jenis_model.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    evaluasi[nama]={
        'prediksi' : y_pred,
        'nilai_true' : y_test,
        'confusion_matrix' : confusion_matrix (y_test,y_pred),
        'akurasi' : accuracy_score (y_test, y_pred),
        'presisi' : precision_score (y_test, y_pred),
        'recall' : recall_score (y_test, y_pred),
        'f1' : f1_score (y_test, y_pred)
    }
```

```python
#Menampilkan prediksi dari semua model algoritma
for nama , hasil in evaluasi.items():
    print(f"\nPrediksi nilai True Model : {nama}")
    print(np.concatenate((hasil['prediksi'].reshape(len(hasil['prediksi']),1),
    hasil['nilai_true'].reshape(len(hasil['nilai_true']),1)),axis=1))
```

```python
#Membuat visual diagram pada banyak model algoritma
import matplotlib.pyplot as plt

def visual(nama,hasil):
    print("\n")
    print(f"\nNama Model : {nama}")
    print(f"Confussion_matrix")
    print(hasil['confusion_matrix'])
    print(f"\nAccuracy : {hasil['akurasi']:.3f}")
    print(f"Precission : {hasil['presisi']:.3f}")
    print(f"Recall     : {hasil['recall']:.3f}")
    print(f"f1 Score   : {hasil['f1']:.3f}")

    label = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
    nilai = [hasil['akurasi'],hasil['presisi'],hasil['recall'],hasil['f1']]

    for index, value in enumerate(nilai):
        plt.text(value,index,str("{:.3f}").format(value))
    plt.barh(label,nilai)
    plt.title(f"Evaluasi Model : {nama}")
    plt.show()
```

```python
#Menampilkan Visual diagram pada banyak model algoritma
for nama, hasil in evaluasi.items():
    visual(nama, hasil)
```

```python
#Menampilkan visual diagram pada 1 model algoritma

visual('Logistic Regression',evaluasi['Logistic Regression'])
```

### PERTEMUAN 5

### Teknik Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

kf = KFold(n_splits=10)
model = LogisticRegression()

kf_accuracies = []
kf_precisions = []
kf_recalls = []
kf_f1s = []
fold = 1
```

```python
for train_index, test_index in kf.split(X):
    X_train_kf, X_test_kf = X[train_index], X[test_index]
    y_train_kf, y_test_kf = y[train_index], y[test_index]

    model.fit(X_train_kf, y_train_kf)
    y_pred_kf = model.predict(X_test_kf)

    acc = accuracy_score(y_test_kf, y_pred_kf)
    precision = precision_score(y_test_kf, y_pred_kf, average='weighted', zero_division=1)
    recall = recall_score(y_test_kf, y_pred_kf, average='weighted', zero_division=1)
    f1 = f1_score(y_test_kf, y_pred_kf, average='weighted', zero_division=1)

    kf_accuracies.append(acc)
    kf_precisions.append(precision)
    kf_recalls.append(recall)
    kf_f1s.append(f1)

    print(f'Fold {fold} - Accuracy: {acc:.3f}, Precision: {precision:.3f}, Recall: {recall:.3f}, F1 Score: {f1:.3f}')
    fold += 1

print(f'\nAverage Accuracy (KFold): {np.mean(kf_accuracies):.3f}')
print(f'Average Precision  (KFold): {np.mean(kf_precisions):.3f}')
print(f'Average Recall     (KFold): {np.mean(kf_recalls):.3f}')
print(f'Average F1 Score   (KFold): {np.mean(kf_f1s):.3f}')
```

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

loo = LeaveOneOut()
model = LogisticRegression()

loo_accuracies = []
loo_precisions = []
loo_recalls = []
loo_f1s = []
fold = 1
```

```python
for train_index, test_index in loo.split(X): # lebih ekstrim, untuk data testingnya satu persatu beda dengan KFol dikelompokkan
    X_train_loo, X_test_loo = X[train_index], X[test_index]
    y_train_loo, y_test_loo = y[train_index], y[test_index]

    model.fit(X_train_loo, y_train_loo)
    y_pred_loo = model.predict(X_test_loo)

    acc       = accuracy_score(y_test_loo, y_pred_loo)
    precision = precision_score(y_test_loo, y_pred_loo, average='weighted', zero_division=1)
    recall    = recall_score(y_test_loo, y_pred_loo, average='weighted', zero_division=1)
    f1        = f1_score(y_test_loo, y_pred_loo, average='weighted', zero_division=1)

    loo_accuracies.append(acc)
    loo_precisions.append(precision)
    loo_recalls.append(recall)
    loo_f1s.append(f1)

    print(f'Fold {fold} - Accuracy: {acc:.3f}, Precision: {precision:.3f}, Recall: {recall:.3f}, F1 Score: {f1:.3f}')
    fold += 1

print(f'\nAverage Accuracy (KFold): {np.mean(loo_accuracies):.3f}')
print(f'Average Precision  (KFold): {np.mean(loo_precisions):.3f}')
print(f'Average Recall     (KFold): {np.mean(loo_recalls):.3f}')
print(f'Average F1 Score   (KFold): {np.mean(loo_f1s):.3f}')
```

### PERTEMUAN 6

### Metode Ensemble

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

def method_ensemble(model, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    evaluasi = {
        'Prediksi': y_pred,
        'Nilai True': y_test,
        'confusion_matrix': confusion_matrix(y_test, y_pred),
        'akurasi': accuracy_score(y_test, y_pred),
        'presisi': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1': f1_score(y_test, y_pred)
    }
    visual(model_name, evaluasi)
    
    return evaluasi
```

```python
# Voting Classifier  

from sklearn.ensemble import VotingClassifier

vc = VotingClassifier(estimators=[
    ('Logistic Regression' , lr),
    ('Random Forest', rf),
    ('Decision Tree', dt),
    ('K-NN', knn)
], voting='soft')

voting = method_ensemble(vc, "Ensemble Voting Classifier (soft)")
```

```python
# Stacking Classifier
from sklearn.ensemble import StackingClassifier 
from sklearn.linear_model import LogisticRegression

sc = StackingClassifier (estimators=[
    ('Logistic Regression', lr), 
    ('Random Forest', rf),
    ('K-NN', knn)
], final_estimator=LogisticRegression())

stacking = method_ensemble (sc, "Ensemble Stacking Classifier")
```

```python
# Boosting Classifier
#1. AdaBoost: Fokus pada memperbaiki kesalahan model tradisional dengan memberi bobot lebih pada data yang sulit diprediksi.
#2. Gradient Boosting: Menambahkan model secara bertahap untuk mengurangi kesalahan prediksi secara keseluruhan.
#3. XGBoost: Implementasi Gradient Boosting yang dioptimalkan untuk kecepatan dan performa tinggi.

from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()

boosting = method_ensemble(gb, "Ensemble Boosting Classifier")
```

```python
# Hybrid Ensemble Classifier
from sklearn.ensemble import BaggingClassifier, GradientBoostingClassifier

gb = GradientBoostingClassifier()
m_bc = BaggingClassifier (estimator=gb)

hybrid_ensemble = method_ensemble (m_bc, "Ensemble Hybrid Classifier")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from tqdm import tqdm

# metrics
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

jenis_model = {
    'RandomForest' : RandomForestClassifier(),
    'SVM' : SVC()
}

parameter = {
    'RandomForest' : {
        'n_estimators' : [10, 50],
        'max_depth' : [10, 20],
        'min_samples_split' : [2, 5]
    },
    'SVM' : {
        'C' : [0.1, 1, 10],
        'kernel' : ['linear', 'rbf'],
        'gamma' : ['scale', 'auto']
    }
}

grid_search_list = []

for model_name, model in jenis_model.items():
    params = parameter[model_name]
    
    grid_search = GridSearchCV(estimator=model, param_grid=params, cv=3, n_jobs=-1, verbose=0, scoring='accuracy')
    
    total_combinations = len(params[list(params.keys())[0]]) * len(params[list(params.keys())[1]]) * len(params[list(params.keys())[2]])
    
    padded_model_name = f"{model_name:<15}"
    
    with tqdm(total = total_combinations * 3, desc = f"Training {padded_model_name}") as pbar :
        grid_search.fit(X_train, y_train)
        pbar.update(1)
    
    grid_search_list.append((model_name, grid_search))
    
for model_name, grid_search in grid_search_list :
    print(f"\nModel : {model_name}")
    print(f"Hyperparamater Terbaik : {grid_search.best_params_}")
    print(f"Skor Terbaik : {grid_search.best_score_:.4f}")
          
    y_pred = grid_search.best_estimator_.predict(X_test)
    accuracy - accuracy_score(y_test, y_pred)
    print(f"Akurasi : {accuracy:.4f}\n")
```
